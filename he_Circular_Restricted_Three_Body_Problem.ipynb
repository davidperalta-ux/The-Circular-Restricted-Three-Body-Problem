```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp


mu = 0.012151 # para el tierra-luna (ver apendice)


xL4 = 0.5 - mu
yL4 = np.sqrt(3)/2



def r1(x, y):
    return np.sqrt((x + mu)**2 + y**2)

def r2(x, y):
    return np.sqrt((x - 1 + mu)**2 + y**2)




def dUdx(x, y):

    termino1 = x

    termino2 = (1 - mu)*(x + mu)/r1(x, y)**3

    termino3 = mu*(x - 1 + mu)/r2(x, y)**3

    return termino1 - termino2 - termino3


def dUdy(x, y):

    termino1 = y

    termino2 = (1 - mu)*y/r1(x, y)**3

    termino3 = mu*y/r2(x, y)**3

    return termino1 - termino2 - termino3



def pcr3bp(t, estado):

    x  = estado[0]
    y  = estado[1]
    vx = estado[2]
    vy = estado[3]

    ax = 2*vy + dUdx(x, y)


    ay = -2*vx + dUdy(x, y)


    return [vx, vy, ax, ay]



x0 = xL4 + 0.005
y0 = yL4

vx0 = 0.0
vy0 = 0.005

state0 = [x0, y0, vx0, vy0]



t0 = 0
tf = 400

t_eval = np.linspace(t0, tf, 40000)


sol = solve_ivp(
    pcr3bp,
    [t0, tf],
    state0,
    t_eval=t_eval,
    rtol=1e-10,
    atol=1e-10
)




x = sol.y[0]
y = sol.y[1]



plt.figure(figsize=(8, 5))

plt.plot(x, y, linewidth=0.8, label="Trajectory around L4")
plt.plot(xL4, yL4, "ko", markersize=5, label="$L_4$")

plt.xlabel("$x$")
plt.ylabel("$y$")
plt.title("Quasi-periodic motion around $L_4$")

plt.xlim(0.0, 0.7)
plt.ylim(0.73, 0.98)

plt.grid(True)
plt.legend()

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt


mu = 0.012151 # para el tierra-luna (ver apendice)


def r1(x, y):
    return np.sqrt((x + mu)**2 + y**2)

def r2(x, y):
    return np.sqrt((x - 1 + mu)**2 + y**2)


def U(x, y):
    return 0.5*(x**2 + y**2) + (1 - mu)/r1(x, y) + mu/r2(x, y)


x = np.linspace(-1.8, 1.8, 1200)
y = np.linspace(-1.8, 1.8, 1200)

X, Y = np.meshgrid(x, y)

Z = 2*U(X, Y)



niveles = [2.9, 3.0, 3.1, 3.17, 3.2, 3.4]


plt.figure(figsize=(8, 8))

plt.contourf(X, Y, Z, levels=niveles, cmap="Greys")
plt.contour(X, Y, Z, levels=niveles, colors="black", linewidths=0.8)

plt.plot(-mu, 0, "ko", markersize=5)
plt.plot(1 - mu, 0, "ko", markersize=5)

plt.xlabel("$x$")
plt.ylabel("$y$")
plt.title("Zero Velocity Curves")


plt.xlim(-1.8, 1.8)
plt.ylim(-1.8, 1.8)

plt.grid(True)
plt.show()
```
